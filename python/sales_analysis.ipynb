```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib as plt

print("Libraries imported successfully!")
```

```python
# Create sample data (in case you don't have the CSV yet)
data = {
    'Month': ['January', 'February', 'March', 'April', 'May', 'June'],
    'Product_A': [120, 135, 140, 155, 160, 170],
    'Product_B': [150, 145, 160, 170, 165, 175],
    'Product_C': [90, 95, 100, 105, 110, 115]
}

# Convert to DataFrame and save as CSV
df = pd.DataFrame(data)
df.to_csv('sales_data.csv', index=False)

print("Sample CSV file created as 'sales_data.csv'")
df  # Display the DataFrame
```

```python
# Load the CSV file
df = pd.read_csv('sales_data.csv')

# Display first 5 rows
df.head()
```

```python
# Add a Total_Sales column
df['Total_Sales'] = df['Product_A'] + df['Product_B'] + df['Product_C']

# Display the modified DataFrame
print("DataFrame with Total Sales:")
df
```

```python
# Calculate average sales using numpy
avg_sales = np.mean(df[['Product_A', 'Product_B', 'Product_C']], axis=0)

print("Average Sales per Product:")
print(f"Product A: {avg_sales[0]:.2f}")
print(f"Product B: {avg_sales[1]:.2f}")
print(f"Product C: {avg_sales[2]:.2f}")

# Bonus: Calculate monthly growth rate using numpy
growth_rates = np.diff(df['Total_Sales']) / df['Total_Sales'][:-1] * 100
print("\nMonthly Growth Rates (%):", np.round(growth_rates, 2))
```

```python
# Import matplotlib.pyplot as plt if not already done
import matplotlib.pyplot as plt

# Create a plot
plt.figure(figsize=(10, 5))

# Plot each product's sales
plt.plot(df['Month'], df['Product_A'], 'b-o', label='Product A')
plt.plot(df['Month'], df['Product_B'], 'r--s', label='Product B')
plt.plot(df['Month'], df['Product_C'], 'g-.D', label='Product C')

# Add labels and title
plt.title("Monthly Sales Performance", fontsize=14)
plt.xlabel("Month", fontsize=12)
plt.ylabel("Units Sold", fontsize=12)
plt.legend(fontsize=10)

# Rotate x-axis labels and add grid
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.7)

# Display the plot
plt.show()
```

```python
# Set position for bars
x = np.arange(len(df['Month']))  # [0,1,2,3,4,5]
width = 0.25  # Width of each bar

# Create bars
plt.figure(figsize=(10, 5))
plt.bar(x - width, df['Product_A'], width, label='Product A')
plt.bar(x, df['Product_B'], width, label='Product B')
plt.bar(x + width, df['Product_C'], width, label='Product C')

# Add details
plt.title("Monthly Sales Comparison", fontsize=14)
plt.xlabel("Month", fontsize=12)
plt.ylabel("Units Sold", fontsize=12)
plt.xticks(x, df['Month'], rotation=45)
plt.legend()

# Display
plt.show()
```
